### 存储成mat 格式

```python

## import 读取对应的包
import pandas as pd
import os
import scipy.io as scio
import numpy as np
from tqdm import tqdm
### label 数据进行读入

full_df = pd.read_csv('../../full_data.csv',index_col= 0)

# full_df ## 根据phq-score 来计算得分

### 读入实验特征数据

au_data_dir = '../../au_data_csv/'

au_data_file = [i for i in os.listdir(au_data_dir)] ## 实验数据路径

for example_data_dir in tqdm(au_data_file):
## 以第一个为例子


    example_data = pd.read_csv(au_data_dir + example_data_dir, index_col= 0 )

    save_dir = '../../au_data_csv_preprocess/'


    ## 寻找对应特征 PoseTx-AU45_r [4:35]

    feature = example_data.iloc[:,4:35].values

    ## 按照30长度均等分
    select_len = 30

    sample_num = feature.shape[0] // select_len

    if feature.shape[0] % select_len > 0:
        feature = feature[:sample_num * select_len,:]

    feature_split = np.split(feature, sample_num, axis= 0 )

    example_label = full_df[full_df['Participant_ID'] == int(example_data_dir.split('_')[0])].PHQ_Score.values

    for idx, save_data in enumerate(feature_split):
        save_path = save_dir + example_data_dir.split('_')[0] + '/'

        if not os.path.exists(save_path):
            os.makedirs(save_path)

        scio.savemat(save_path + str(idx) + '.mat', {'feature':save_data,'label':example_label})


```
